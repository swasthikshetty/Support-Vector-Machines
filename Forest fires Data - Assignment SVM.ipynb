```python
import pandas as pd 
import numpy as np 
import seaborn as sns
```

```python
data = pd.read_csv("C:\\Users\\Swathi\\Desktop\\Excelr Classes\\Assignments\\Support Vector Machines\\forestfires.csv")
data.head()
# data.describe()
# data.columns
```

```python
data = data.drop(['month'], axis=1)
data = data.drop(['day'], axis=1)
data.head()
```

```python
# sns.pairplot(data=letters)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['size_category'] = le.fit_transform(data['size_category'].astype(str))
data.head()
```

```python
data['size_category'].value_counts()
```

```python
data.shape
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
train,test = train_test_split(data,test_size = 0.2)
train_X = train.iloc[:,:28]
train_y = train.iloc[:,28]
test_X = test.iloc[:,:28]
test_y = test.iloc[:,28]
```

# Create SVM classification object 

# 1. Kernal = linear

```python
model_linear = SVC(kernel = "linear")
model_linear.fit(train_X,train_y)
pred_test_linear = model_linear.predict(test_X)

np.mean(pred_test_linear==test_y) # Accuracy = 100
```

# 2. Kernel = poly

```python
model_poly = SVC(kernel = "poly")
model_poly.fit(train_X,train_y)
pred_test_poly = model_poly.predict(test_X)

np.mean(pred_test_poly==test_y) # Accuracy = 79.80
```

# 3. kernel = rbf

```python
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(train_X,train_y)
pred_test_rbf = model_rbf.predict(test_X)

np.mean(pred_test_rbf==test_y) # Accuracy = 77.88
```
